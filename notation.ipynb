```python
df = pd.read_csv('/content/drive/MyDrive/Sounding_Spirit/notation.csv')
df["ssdl:notation"] = df["ssdl:notation"].astype(str)
df.head()
```

```python
# plot the number of books using each notation
import numpy as np
import matplotlib.pyplot as plt

data = {}
for x in df.index:
  for notation in df.loc[x,"ssdl:notation"].split(";"):
    if notation not in data:
      data[notation]=0
    data[notation] = data[notation]+1

data.pop("Nil")
data.pop("nan")

# print(data)
data = dict(sorted(data.items(), key=lambda item: item[1]))

notations = list(data.keys())
values = list(data.values())

fig, ax = plt.subplots()
fig.set_figwidth(10)

# Example data
y_pos = np.arange(len(notations))

ax.barh(y_pos, values, align='center')
ax.set_yticks(y_pos, labels=notations)
ax.set_xlabel('Number of songbooks')
ax.set_title('Number of books with different Notation values')

for i in ax.patches:
    plt.text(i.get_width()+1, i.get_y()+0.25,
             str(round((i.get_width()), 2)),
             fontsize = 10,
             color ='black')

plt.show()
```

```python
# all seven shape notations are merged together
# all numberal notations are merged together
import numpy as np
import matplotlib.pyplot as plt

data = {}
for x in df.index:
  for notation in df.loc[x,"ssdl:notation"].split(";"):
    if notation.startswith("Seven-shape"):
      notation = "Seven-shape"
    if notation.startswith("Numeral"):
      notation = "Numeral"
    if notation not in data:
      data[notation]=0
    data[notation] = data[notation]+1

data.pop("Nil")
data.pop("nan")

# print(data)
data = dict(sorted(data.items(), key=lambda item: item[1]))

notations = list(data.keys())
values = list(data.values())

fig, ax = plt.subplots()
fig.set_figwidth(10)

# Example data
y_pos = np.arange(len(notations))

ax.barh(y_pos, values, align='center')
ax.set_yticks(y_pos, labels=notations)
ax.set_xlabel('Number of songbooks')
ax.set_title('Number of books with different Notation values')

for i in ax.patches:
    plt.text(i.get_width()+1, i.get_y()+0.25,
             str(round((i.get_width()), 2)),
             fontsize = 10,
             color ='black')

plt.show()
```

```python
# plot the number of books using each notation over years
import numpy as np
import matplotlib.pyplot as plt

data = {}
for x in df.index:
  for notation in df.loc[x,"ssdl:notation"].split(";"):
    if notation not in data:
      data[notation]={}

    # excluded all books with unclear publishing dates
    if not df.loc[x,"ssdl:pid"][:4].isnumeric():
      continue

    year = int(df.loc[x,"ssdl:pid"][:4])
    # print(year,df.loc[x,"ssdl:pid"])
    if year not in data[notation]:
      data[notation][year]=0
    data[notation][year]=data[notation][year]+1

print(data)

data.pop("Nil")
data.pop("nan")

plt.figure(figsize=(10,6))

# print(data)
for notation in data:
  data[notation] = dict(sorted(data[notation].items(), key=lambda item: item[0]))
  plt.plot(list(data[notation].keys()),list(data[notation].values()),label=notation)

plt.legend(loc="upper left")
plt.show()
```
